```python
import l2f_tda as tda
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
    
%matplotlib inline
```

```python
import pickle as pkl
df_series = pkl.load(open('example_time_series.pkl', 'rb'))
df_regime = pkl.load(open('example_regime.pkl', 'rb'))
```

```python
fig = plt.figure(figsize=(10,4))

plt.plot(df_series)
plt.plot(10*df_regime, color='r')
print(df_series.shape)
```

```python
outerWindowDuration = 100
outerWindowStride = 10
innerWindowDuration = 100
innerWindowStride = 10
series_embedder = tda.TakensEmbedder(outerWindowDuration, outerWindowStride, innerWindowDuration, innerWindowStride)
```

```python
series_embedder.fit(df_series)
df_embedded_series = series_embedder.transform(df_series)
print(df_embedded_series.shape)
```

```python
outerWindowDuration = 100
outerWindowStride = 10
innerWindowDuration = outerWindowDuration
innerWindowStride = 1
regime_embedder = tda.TakensEmbedder(outerWindowDuration, outerWindowStride, innerWindowDuration, innerWindowStride)
```

```python
regime_embedder.fit(df_regime)
df_embedded_regime = regime_embedder.transform(df_regime)
print(df_embedded_regime.shape)
```

```python
fig = plt.figure(figsize=(10,4))

plt.plot(df_embedded_series[0,0,:])
plt.plot(10*df_embedded_regime[0,0,:], color='r')
```
